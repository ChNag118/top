```python
import random
from datetime import datetime
import numpy as np
def random_date_generator(start_date, range_in_days):
    days_to_add = np.arange(0, range_in_days)
    random_date = np.datetime64(start_date) + np.random.choice(days_to_add)
    return random_date
delivery = pd.DataFrame(columns=['Delivery Id','Purchase Order ID','Ship Date','Shipped Qty','Planned delivery Date','Actual Delivery Date'])
for i in range(1000):
    item = df_del_5.sample()
    item1=df_pur_5.sample()
    ship_qty= str(random.randint(1,100))
    booked_date=datetime.now()
    olo = random_date_generator('2021-10-19', 30)
    olo1 = random_date_generator('2021-11-19', 30)
    olo2 = random_date_generator('2021-12-19', 30)
    time = booked_date.strftime("%d/%m/%Y")
    booked_time_hours=booked_date.strftime('%H:%M:%S')
    delivery_id= "DE"+str(random.randint(1,100))+ pd.util.testing.rands(4)
    pur_id= "PO"+str(random.randint(1,100))+ pd.util.testing.rands(4)
    delivery.loc[len(delivery.index)] = [delivery_id,pur_id,olo,ship_qty,olo1,olo2] 
delivery
```
